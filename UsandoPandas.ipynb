```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = {

        'Nome': ['Alice','Bob','Charlie','David','Emily'],
        'Idade': [25,30,35,40,27],
        'Cargo': ['Analista','Gerente','CEO','Analista','Coordenador'],
         'Salario': [50000,80000,15000,48000,6000]

}

```

```python

df = pd.DataFrame(data)
df
```

```python
df[df['Salario'] > 8000]
```

```python
df.sort_values(by='Idade', ascending=False)
```

```python
df['Bônus'] = df['Salario'] * 0.1
df
```

```python
df.groupby('Cargo')['Salario'].mean()
```

```python
df.plot(
    kind='bar',
    x='Nome',
    y='Salario',
    title='Salário dos Funcionarios',
    rot = 45



)
```
