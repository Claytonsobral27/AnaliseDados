```python
import pandas as pd
import openpyxl as xl

```

```python
dados = {

   'Times':['São paulo','Palmeiras', 'Flamengo','Santos','Paysandu' ],
   'Partidas_jogadas':[10,10,10,10,10],
   'Vitorias':[7,5,6,4,3],
   'Empates':[2,2,2,1,1],
   'Derrtotas':[1,3,4,5,6],
   'Gols_Pro':[22,18,16,14,10],
   'Gols_Contra':[10,12,15,16,20]



}


df = pd.DataFrame(dados)
df
```

```python
df['Saldo_Gols'] = df['Gols_Pro'] - df['Gols_Contra']
df
```

```python
df['Pontos'] = df['Vitorias'] * 3 + df['Empates'] * 1
df
```

```python


df = df.sort_values(by='Pontos' , ascending=False)
df
```

```python
df.to_excel('campeonato_Brasileiro.xlsx')
```
