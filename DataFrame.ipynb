```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
dados_vendas = {

                'produtos': ['Camisa', 'Calça', 'Saia','Camisa','Calça','Saia','Camisa','Calça','Saia'],
                'Preço' : [50,100,80,55,110,85,45,95,90],
                'Quantidade':[10,15,20,5,10,15,30,20,10],
                'Mês':['Jan','Jan','Jan','Fev','Fev','Fev','Mar','Mar','Mar']



}
```

```python
df_vendas = pd.DataFrame(dados_vendas)
df_vendas
```

```python
df_vendas['Receita'] = df_vendas['Preço'] * df_vendas['Quantidade']
df_vendas
```

```python
# Sumarizando a receita total do produto
resumo_vendas =df_vendas.groupby('produtos')['Receita'].sum().reset_index()
resumo_vendas
```

```python
produtos_altas_vendas = resumo_vendas[resumo_vendas['Receita'] > 1000]
produtos_altas_vendas
```

```python
plt.figure(figsize=(8,5))
plt.bar(resumo_vendas['produtos'],resumo_vendas['Receita'],color='skyblue')
plt.title('Receita Total por Produto')
plt.xlabel('Produtos')
plt.ylabel('Receita')
plt.xticks(rotation=45)
plt.show()

```
